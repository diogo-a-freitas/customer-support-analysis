## Convert Order Time Stamps to Date Time and Feature Engineer

```python
orders.describe()
```

```python
from datetime import datetime

orders[['order_purchase_timestamp', 'order_approved_at', \
        'order_delivered_carrier_date', 'order_delivered_customer_date', \
        'order_estimated_delivery_date']] = orders[['order_purchase_timestamp', 'order_approved_at', \
        'order_delivered_carrier_date', 'order_delivered_customer_date', \
        'order_estimated_delivery_date']].apply(pd.to_datetime, errors='coerce')
```

### CUSTOMER ORDER FEATURES

```python
# week number

orders['order_week_purchase'] = orders['order_purchase_timestamp'].dt.strftime('%W')

#month of purchase
orders['order_month_purchase'] = orders['order_purchase_timestamp'].dt.strftime('%m')

#year of purchase
orders['order_year_purchase'] = orders['order_purchase_timestamp'].dt.strftime('%Y')

#day of week of purchase
orders['order_dow_purchase'] = orders['order_purchase_timestamp'].dt.strftime('%a')

#time of purchase
orders['order_time_purchase'] = orders['order_purchase_timestamp'].dt.strftime('%X')
```

```python
#function to apply date mapping

orders['order_time_day_purchase'] = orders['order_time_purchase'].apply(
    lambda x: 'morning' if x >= '06:00:00' and x < '12:00:00' else
              'afternoon' if x >= '12:00:00' and x < '18:00:00' else
              'evening' if x >= '18:00:00' and x < '21:00:00' else
              'night'
)
```

### APPROVED ORDER FEATURES

```python
orders['order_week_approved'] = orders['order_approved_at'].dt.strftime('%W')

#month of approved
orders['order_month_approved'] = orders['order_approved_at'].dt.strftime('%m')

#year of approved
orders['order_year_approved'] = orders['order_approved_at'].dt.strftime('%Y')

#day of week of approved
orders['order_dow_approved'] = orders['order_approved_at'].dt.strftime('%a')

#time of approved
orders['order_time_approved'] = orders['order_approved_at'].dt.strftime('%X')
```

### Carrier Delivered Features

```python
orders['order_week_carrier'] = orders['order_delivered_carrier_date'].dt.strftime('%W')

#month of carrier
orders['order_month_carrier'] = orders['order_delivered_carrier_date'].dt.strftime('%m')

#year of carrier
orders['order_year_carrier'] = orders['order_delivered_carrier_date'].dt.strftime('%Y')

#day of week of carrier
orders['order_dow_carrier'] = orders['order_delivered_carrier_date'].dt.strftime('%a')

#time of carrier
orders['order_time_carrier'] = orders['order_delivered_carrier_date'].dt.strftime('%X')
```

### DELIVERED TO CUSTOMER FEATURES

```python
orders['order_week_customer_received'] = orders['order_delivered_customer_date'].dt.strftime('%W')

#month of customer received
orders['order_month_customer_received'] = orders['order_delivered_customer_date'].dt.strftime('%m')

#year of customer received
orders['order_year_customer_received'] = orders['order_delivered_customer_date'].dt.strftime('%Y')

#day of week of customer received
orders['order_dow_customer_received'] = orders['order_delivered_customer_date'].dt.strftime('%a')

#time of customer received
orders['order_time_customer_received'] = orders['order_delivered_customer_date'].dt.strftime('%X')
```

### Estimated Receive Time

```python
orders['order_week_customer_received_estimate'] = orders['order_estimated_delivery_date'].dt.strftime('%W')

#month of customer received estimate
orders['order_month_customer_received_estimate'] = orders['order_estimated_delivery_date'].dt.strftime('%m')

#year of customer received estimate
orders['order_year_customer_received_estimate'] = orders['order_estimated_delivery_date'].dt.strftime('%Y')

#day of week of customer received estimate
orders['order_dow_customer_received_estimate'] = orders['order_estimated_delivery_date'].dt.strftime('%a')
```

## Calculate time differences

```python
from datetime import datetime, timedelta
#time delta.days gives integer of difference in days, so we need to incorportate the seconds to have it in float
# 1 days = 86400.0 so we convert the difference to seconds, and then divide it by the total number of seconds in a day

def calculate_time(dataframe):
    
    dataframe['time_diff_purchased_received'] = dataframe.apply(
        lambda x: (x['order_delivered_customer_date'] - x['order_purchase_timestamp']).total_seconds() / 86400
        if x['order_delivered_customer_date'] is not None else 'Not Delivered Yet', axis=1
    )
    
    dataframe['time_diff_purchased_received_estimate'] = dataframe.apply(
        lambda x: (x['order_delivered_customer_date'] - x['order_estimated_delivery_date']).total_seconds() / 86400
        if x['order_delivered_customer_date'] is not None else 'Not Delivered Yet', axis=1
    )
    
    dataframe['received_before_after'] = dataframe.apply(
        lambda x: 1 if x['time_diff_purchased_received_estimate'] <= 0 else 0 
        if isinstance(x['time_diff_purchased_received_estimate'], float) else None, axis=1
    )
    
    dataframe['time_diff_purchased_approved'] = dataframe.apply(
        lambda x: (x['order_approved_at'] - x['order_purchase_timestamp']).total_seconds() / 86400
        if x['order_approved_at'] is not None else 'Not Delivered Yet', axis=1
    )
    
    return dataframe['time_diff_purchased_received'], dataframe['time_diff_purchased_received_estimate'], \
                            dataframe['received_before_after']
```

```python
#apply formula 

calculate_time(orders)
```
